```python
import matplotlib.pyplot as plt
import math
```

```python
def plot_fair(name, data):
    x = [point[0] for point in data]
    y = [point[1] for point in data]

    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, color='black')

    plt.axvline(x=0.2, color='black', linestyle='--')

    plt.xlim(0, 1.0)
    plt.xticks([i*0.1 for i in range(11)])

    ax = plt.gca()
    ax.yaxis.tick_right()
    ax.yaxis.set_tick_params(pad=-10)
    for label in ax.get_yticklabels():
        label.set_ha('right')
    ax.tick_params(axis='both', direction='in')

    for i, txt in enumerate(x):
        if txt < 0.0001:
            precision = -math.floor(math.log10(abs(txt)))
            txt = f'{txt:.{precision}f}'
        else:
            txt = str(txt)
        plt.text(x[i], y[i], txt, color='black', fontsize=12, ha='left', va='bottom')

    # plt.title('Fairness Measures of ProPublica COMPAS dataset\nPrivileged Group: African-American', loc='center')

    plt.savefig(f"{name}.pdf", bbox_inches='tight')
```

```python
data = [(0.26, 'demographic parity'),
        (0.25, 'conditional statistical parity'),
        (0.02, 'overall accuracy eqaulity'),
        (0.26, 'mean difference'),
        (0.23, 'equalized odds(true positive)\nequal opportunity'),
        (0.22, 'equalized odds(false positive)\npredictive equality'),
        (0.06, 'conditional use accuracy equality(true positive)\npredictive parity'),
        (0.06, 'conditional use accuracy equality(true negative)'),
        (0.03, 'equal calibration'),
        ]
data.reverse()
plot_fair("African-American", data)
```

```python
def plot_balacne(name, data):
    x = [point[0] for point in data]
    y = [point[1] for point in data]

    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, color='black')

    plt.xlim(0, 10)
    plt.xticks([i for i in range(11)])

    plt.ylim(-0.5, len(y)-0.5)

    ax = plt.gca()
    ax.yaxis.tick_right()
    ax.yaxis.set_tick_params(pad=-10)
    for label in ax.get_yticklabels():
        label.set_ha('right')
    ax.tick_params(axis='both', direction='in')

    # plt.axvline(x=5, color='black', linestyle='--')

    for i, txt in enumerate(x):
        if txt < 0.0001:
            precision = -math.floor(math.log10(abs(txt)))
            txt = f'{txt:.{precision}f}'
        else:
            txt = str(txt)
        plt.text(x[i], y[i], txt, color='black', fontsize=12, ha='left', va='bottom')

    # plt.title('Fairness Measures of ProPublica COMPAS dataset\nPrivileged Group: African-American', loc='center')

    # Show plot
    plt.savefig(f"{name}.pdf", bbox_inches='tight')
```
